# Normalization Vs Standardization

```python
import pandas as pd
```

```python
df = pd.read_csv('500hits.csv', encoding = 'latin-1')
df.head()
```

```python
df = df.drop(columns = ['PLAYER', 'CS'])
```

```python
df.info()
```

```python
df.describe().round(3)
```

```python
# for Strandard_Scalar
X1 = df.iloc[: ,: 14]

# For MinMaxScalar

X2 = df.iloc[:, : 14]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X1 = sc.fit_transform(X1)
X1
```

```python
# Now we convert this data back to dataframe
dataframe = pd.DataFrame(X1, columns = ['YRS','G','AB','R','H','2B','3B','HR','RBI','BB','SO','SB','BA','HOF'])
dataframe.head()
```

```python
dataframe.describe().round(3)
```
