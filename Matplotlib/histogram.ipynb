```python
import matplotlib.pyplot as plt
```

```python
#creating figures
fig = plt.figure(figsize=(12, 5))
# show the plot
figure = plt.figure()
```

```python
# creating plot
# line plot

X =[1,2,3,4,5]
Y =[2,5,7,8,10]
plt.plot(X,Y)
plt.show()
```

```python
# costimizing plot
# creating plot

X =[1,2,3,4,5]
Y =[2,5,7,8,10]
plt.title("line plot", fontsize=20)
plt.plot(X,Y)
plt.show()
```

```python
# creating plot
# line plot

X =[1,2,3,4,5]
Y1 =[2,5,7,8,10]
Y2 =[1,3,5,7,9]

plt.plot(X, Y1, label="line 1")
plt.plot(X, Y2, label="line 2")

# plt.grid="true"

plt.legend()

plt.plot(X,Y)

plt.show()
```
