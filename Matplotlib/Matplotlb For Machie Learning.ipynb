# Stack Plot or Area Plot

```python
# multipple plot in one diagram is called 
# it is also called area graph because it represent area 

import matplotlib.pyplot as plt
a = [1, 2, 3 ,4 ,5 ,6]
area = [1, 2, 4, 2, 5, 1]
area1 = [1, 4, 4, 2, 5, 1]
area2 = [1, 5, 4, 2, 5, 1]

labels = ['area', 'area2', 'area3']

plt.figure(figsize = (5, 4))
plt.stackplot(a, area,area1 , area2 , labels= label, colors = ['r','g','b'], baseline = 'sym' )
plt.legend(loc = 2)
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [1, 2, 3 ,4 ,5 ,6]
area = [1, 2, 4, 2, 5, 1]
area1 = [1, 4, 4, 2, 5, 1]
area2 = [1, 5, 4, 2, 5, 1]

labels = ['area', 'area2', 'area3']

plt.figure(figsize = (5, 4))
plt.title('python')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.grid()
plt.stackplot(a, area,area1 , area2 , labels= label, colors = ['r','g','b'], baseline = 'wiggle' )
plt.legend(loc = 2)
plt.show()
```

# Stepplot

```python
# Diagram look like step 

import matplotlib.pyplot as plt

st1 = [1, 2, 3, 4, 5]
st2 = [1, 2, 3, 4, 5]

plt.figure(figsize = (5, 4))

plt.step(st2, st1, marker = 'o', color = 'r', ms = 10, mfc = 'g', label = ['Python'], )
plt.legend(loc = 2)
plt.title('Python')
plt.xlabel('X_axis')
plt.ylabel('y_axis')
plt.grid()
plt.show()
```

# Fill_Between Plot

```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.figure(figsize = (5, 4))

plt.plot(x, y, color = 'r')

plt.fill_between(x=[2, 4], y1=2, y2=4,  )
plt.title('Python')
plt.xlabel('X_axis')
plt.ylabel('y_axis')
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])

plt.figure(figsize = (5, 4))

plt.plot(x, y, color = 'r')

plt.fill_between(x, y ,where = (x >= 2) & ( x <= 4), alpha = 0.5, )
plt.title('Python')
plt.xlabel('X_axis')
plt.ylabel('y_axis')
plt.grid()

plt.show()
```
