# COUNTPLOT OR CATPLOT

```python
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns

data = sns.load_dataset('tips')
data.head(3)

plt.figure(figsize = (5, 4))
sns.catplot(x = 'day', y = 'size' , data = data , hue = 'sex', palette = 'Oranges', kind = 'bar')
plt.show()
```

# PAIRPLOT

```python
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns

clf = sns.load_dataset('tips').head(20)
clf

sns.pairplot(data, hue = 'sex')
plt.show()
```

# BOXPLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

clf = sns.load_dataset('tips')
clf.head(20)
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'day', y = 'total_bill', data = clf, hue = 'time', color = 'g' ,
            order = ['fFri', 'Sat', 'sSun', 'Thur'], showmeans = True,
            meanprops = {'marker': '+', 'markeredgecolor': 'g'}, linewidth = 3, 
           palette = 'plasma', orient = 'v')

plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(x = 'total_bill', y = 'day', data = clf, hue = 'time', color = 'g' ,
            order = ['fFri', 'Sat', 'sSun', 'Thur'], showmeans = True,
            meanprops = {'marker': '+', 'markeredgecolor': 'g'}, linewidth = 3, 
           palette = 'plasma')

plt.show()
```

```python
sns.set(style = 'whitegrid')

sns.boxplot(data = clf, color = 'g' ,showmeans = True,
            meanprops = {'marker': '+', 'markeredgecolor': 'g'}, linewidth = 3, 
           palette = 'plasma', orient = 'h')

plt.show()
```
