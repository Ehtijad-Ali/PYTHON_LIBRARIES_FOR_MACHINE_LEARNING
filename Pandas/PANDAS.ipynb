```python
# replacing the empty values we used fillna() which replace empty cells with values

import pandas as pd
ehty_new=pd.read_excel('/content/drive/MyDrive/UNCLEAN DATASET.xlsx')
ehty_new.fillna(160, inplace=True)
print(ehty_new)
```

```python
# to replace only empty value for one colunm you need to specified the column name

import pandas as pd
ehty_new=pd.read_excel('/content/drive/MyDrive/UNCLEAN DATASET.xlsx')
ehty_new["CALORIES"].fillna(199, inplace=True)
print(ehty_new.to_string)
```

```python
# here we can also inplace the empty cell using mean(), medium()  and mode

# MEAN
import pandas as pd
ehty_new=pd.read_excel('/content/drive/MyDrive/UNCLEAN DATASET.xlsx')
x= ehty_new["PULSE"].mean()
ehty_new["PULSE"].fillna(x, inplace=True)
print(ehty_new.to_string())
```

```python
# MEDIAN:replace any empty value in it
import pandas as pd
ehty_new=pd.read_excel('/content/drive/MyDrive/UNCLEAN DATASET.xlsx')
x= ehty_new["PULSE"].median()
ehty_new["PULSE"].fillna(x, inplace=True)
print(ehty_new.to_string)
```

```python
#   to calculate mode inpalce empty cell in it
import pandas as pd
ehty_new=pd.read_excel('/content/drive/MyDrive/UNCLEAN DATASET.xlsx')
x= ehty_new["PULSE"].mode()[0]
ehty_new["PULSE"].fillna(x, inplace=True)
print(ehty_new.to_string)
```
