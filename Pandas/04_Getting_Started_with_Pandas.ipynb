Function Application and Mappping

```python
df = pd.DataFrame(np.random.standard_normal((4,3)),
                  columns = list('bde'),
                  index = ['Utah', 'Ohio', 'Texas', 'Oregon'])
df
```

```python
np.abs(df)
```

```python
def f1(x):
  return x.max() - x.min()
```

```python
df.apply(f1)
```

```python
df.apply(f1, axis = 'columns')
```

```python
def f2(x):
  return pd.Series([x.min(), x.max()], index = ['min', 'max'])
```

```python
df.apply(f2)
```

```python
def my_format(x):
  return f"{x:.2f}"
```

```python
df.applymap(my_format)
```

```python
df['e'].map(my_format)
```
